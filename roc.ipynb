```
import tensorflow as tf
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
import numpy as np
```

```
model = tf.keras.models.load_model('/work/Cavity_detection/models/DCNN3/best_model.h5')
```

```
# Load the arrays back into memory
X_test = np.load('/work/Cavity_detection/data/test/X_test.npy')
y_test = np.load('/work/Cavity_detection/data/test/y_test.npy')
```

```
# Predict probabilities for the test set
y_pred = model.predict(X_test)

# Compute the false positive rate, true positive rate, and thresholds for the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)

# Plot the ROC curve
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```
